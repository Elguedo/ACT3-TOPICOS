```python
housing_data = pd.read_csv("/content/drive/MyDrive/ACT3 TOPICOS/Housing.csv", header=0, sep=",")
```

```python
#tamaño del conjunto de datos
housing_data.shape
```

```python
# Se asignan los valores de X y Y para generar el grafico.
# Se escoge la variable X como el area de la casa y la variable Y como los precios de las casa
x = housing_data["area"]
y = housing_data["price"]
```

```python
# Se obtienen algunos valores clave importantes
slope, intercept, r, p, std_err = stats.linregress(x, y)
```

```python
# Se crea una función que utilice los valores de pendiente e intersección para devolver un nuevo valor
def myfunc(x):
 return slope * x + intercept
```

```python
#Se ejecuta cada valor de la matriz x a través de la función myfunc
mymodel = list(map(myfunc, x))
```

```python
# Se genera el grafico de dispersion de la regresion lineal de Precios de casas en función de area
plt.scatter(x, y)
# Se dibuja la linea de regresion lineal
plt.plot(x, slope * x + intercept)
# Se definen los. valores máximos y mínimos de cada eje
plt.ylim(ymin=0, ymax=20000000)
plt.xlim(xmin=0, xmax=20000)
#Se definen las etiquetas de los ejes
plt.xlabel("Area")
plt.ylabel ("Precio")
plt.show()
```

```python
# Se importa la libreria statsmodel.formula.api
import statsmodels.formula.api as smf
```
